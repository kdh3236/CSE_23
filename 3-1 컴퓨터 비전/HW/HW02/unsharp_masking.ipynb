```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mean_filter(size):
    return np.ones((size, size), np.float32) / (size*size)
```

```python
def make_identity_filter(size):
    filter = np.zeros((size, size), np.float32)
    filter[size // 2, size // 2] = 1.0
    return filter
```

```python
def unsharp_masking(size, alpha, image):
    blur_filter = mean_filter(size)
    identity_filter = make_identity_filter(size)
    sharpen_filter = ((1+alpha) * identity_filter) - (alpha * blur_filter)
    return cv2.filter2D(image, -1, sharpen_filter)
```

```python
raw_image = cv2.imread('image.jpg')
image = cv2.cvtColor(raw_image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
sharp_image = unsharp_masking(3, 5, image)
plt.imshow(sharp_image)
```

```python
sharp_image = unsharp_masking(5, 5, image)
plt.imshow(sharp_image)
```

```python
sharp_image = unsharp_masking(7, 5, image)
plt.imshow(sharp_image)
```
